```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = Image.open('2609.jpg')
img.load()
data = np.asarray(img, dtype='int32')

print(data.shape)

outer_border = 87
card_width = 852
card_height = 1566
card_spacer = 63

height, width, _ = data.shape

for height in range(2):
    for width in range(7):
        start_x = outer_border + (card_width + card_spacer) * width
        start_y = outer_border + (card_height + card_spacer) * height
        card = data[start_y:start_y + card_height, start_x:start_x + card_width, :]
        img = Image.fromarray(np.uint8(card)).convert('RGB')
        img = img.resize((142, 261), Image.ANTIALIAS)
        img.save('card{}{}.png'.format(height, width))
        plt.imshow(card)
```
